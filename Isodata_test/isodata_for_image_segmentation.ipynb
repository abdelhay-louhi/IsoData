<h3><span style="color: blue;"> üîç Qu'est-ce que ISODATA ?</span></h3>

"**ISODATA**" signifie "Iterative Self-Organizing Data Analysis Technique Algorithm" (Algorithme de technique d‚Äôanalyse de donn√©es auto-organisatrice it√©rative).
C‚Äôest un algorithme de **clustering non supervis√©** qui est **une extension de l‚Äôalgorithme K-means**, mais** plus flexible et adaptatif**.

<h3><span style="color: blue;">üìå Caract√©ristiques principales :</span></h3>


*   **It√©ratif :** il r√©p√®te le processus jusqu‚Äô√† convergence ou jusqu‚Äô√† atteindre un nombre maximal d‚Äôit√©rations.



*   **Auto-organis√© :** il s‚Äôadapte √† la structure des donn√©es.

*   **Nombre dynamique de clusters :** il peut diviser un cluster s‚Äôil est trop dispers√© ou fusionner deux clusters s‚Äôils sont trop proches.

*   **Adapt√© aux donn√©es de haute dimension.**


<h3><span style="color: blue;">‚öôÔ∏è √âtapes principales de l‚Äôalgorithme ISODATA :</span></h3>

<h5><span style="color: green;">Initialisation :</span></h5>


*   Choisir un nombre initial de clusters (sup√©rieur ou √©gal au nombre attendu).
*   Assigner des centro√Ødes initiaux aux clusters.

<h5><span style="color: green;">Affectation :</span></h5>

*   Assigner chaque point au centro√Øde le plus proche.

<h5><span style="color: green;"> Mise √† jour :</span></h5>

*   Calculer de nouveaux centro√Ødes √† partir de la moyenne des points assign√©s.


<h5><span style="color: green;"> √âvaluation :</span></h5>   


*   **Diviser** les clusters dont la variance d√©passe un certain seuil.

*  **Fusionner** les clusters trop proches.

*   **√âliminer** les clusters contenant trop peu de points.


 <h5><span style="color: green;"> It√©ration :</span></h5>   

*   R√©p√©ter les √©tapes d‚Äôassignation, de mise √† jour et d‚Äô√©valuation jusqu‚Äô√† convergence ou jusqu‚Äôau nombre maximal d‚Äôit√©rations.

<h3><span style="color: blue;">üß† Cas d‚Äôutilisation :</span></h3>


*   Segmentation d‚Äôimages

*   T√©l√©d√©tection
*   Reconnaissance de formes


*   Segmentation de march√©



<h3><span style="color: blue;">Comparaison entre ISODATA et K-Means : Deux approches du clustering, deux fa√ßons de r√©fl√©chir</span></h3>

**K-Means** suppose que le nombre de groupes (clusters) est connu √† l‚Äôavance, et applique un processus fixe et r√©p√©titif jusqu‚Äô√† la stabilisation des r√©sultats.

En revanche, **ISODATA** est plus flexible et intelligent : il commence avec un nombre initial de clusters, puis les fusionne ou les divise au fil des it√©rations selon la dispersion des donn√©es.

Cela permet √† ISODATA de mieux s‚Äôadapter √† la structure r√©elle des donn√©es, surtout quand celle-ci est complexe ou inconnue.

Exemple illustratif : deux √©tudiants face √† un devoir

<h5><span style="color: green;"> l‚Äôexemple de deux √©tudiants,:</span></h5>  
Prenons l‚Äôexemple de deux √©tudiants, Youssef et Amine, confront√©s √† un

devoir difficile.

* **Youssef (K-Means)** suit un plan de r√©solution qu‚Äôil a appris en cours. Il applique les √©tapes sans se poser de questions, m√™me si les r√©sultats sont moyens. Il ne change pas de m√©thode, et continue jusqu‚Äô√† obtenir une r√©ponse stable.

*   **Amine (ISODATA)** adopte une approche plus critique. Apr√®s chaque tentative, il s‚Äôinterroge sur la clart√© et la pertinence de ses d√©marches. Si deux √©tapes lui paraissent redondantes, il les regroupe. Si une partie du raisonnement est confuse, il la divise en sous-parties plus simples. Il ajuste sa m√©thode au fur et √† mesure pour mieux comprendre le probl√®me.

<h5><span style="color: green;"> Conlusion</span></h5>

**Youssef applique une strat√©gie fixe, rapide et directe, comme K-Means. Amine adapte sa m√©thode selon les besoins du probl√®me, comme ISODATA**. **Le choix** entre les deux **d√©pend du contexte** : la **simplicit√© de K-Means** est efficace quand **les objectifs sont clairs**, tandis que **la souplesse d‚ÄôISODATA est pr√©f√©rable** quand **l‚Äôorganisation des donn√©es est incertaine**.

<h5><span style="color: green;">Exemple simplifi√© :</span></h5>
Supposons que tu as 6 pixels (valeurs de gris) que tu veux regrouper en 2 clusters :

**Valeurs des pixels : [10, 12, 11, 90, 92, 89]**
Et ton mod√®le (ex. K-Means) a fait cette s√©paration :

**Cluster 0 : [10, 12, 11]       ‚Üí centre = (10 + 12 + 11)/3 = 11**

**Cluster 1 : [90, 92, 89]       ‚Üí centre = (90 + 92 + 89)/3 = 90.33**

<h5><span style="color: green;">üîπ √âtape 1 : Variance intra-cluster</span></h5>

On calcule la variance pour chaque cluster, puis on prend la moyenne.

**Pour Cluster 0 (valeurs [10, 12, 11], centre = 11) :**

* (10 - 11)¬≤ = 1
* (12 - 11)¬≤ = 1
* (11 - 11)¬≤ = 0

=> Moyenne = (1 + 1 + 0)/3 = 0.67

**Pour Cluster 1 (valeurs [90, 92, 89], centre ‚âà 90.33) :**

*(90 - 90.33)¬≤ ‚âà 0.11
*(92 - 90.33)¬≤ ‚âà 2.78
*(89 - 90.33)¬≤ ‚âà 1.78

=> Moyenne ‚âà (0.11 + 2.78 + 1.78)/3 ‚âà 1.56

Variance intra-cluster globale :

* (0.67 + 1.56)/2 = 1.115

‚û°Ô∏è Plus la variance est faible, plus les groupes sont compacts ‚Üí c‚Äôest mieux !


<h5><span style="color: green;">üîπ √âtape 2 : Distance moyenne inter-centres</span></h5>

Tu as 2 centres : 11 et 90.33

Donc la distance entre eux :

* |11 - 90.33| = 79.33

‚û°Ô∏è Plus la distance entre les centres est grande, plus les clusters sont bien s√©par√©s ‚Üí c‚Äôest mieux !

<h3><span style="color: blue;"> ‚òÖ‚òÖ‚òÖ‚òÖ‚òÖ‚òÖ‚òÖ √âtude comparative de la segmentation d‚Äôimages par K-means et IsoData</span></h3>


```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans
from scipy.spatial.distance import pdist

# üîπ Fonction ISODATA simplifi√©e
def isodata_segmentation(image_array, K_init=3, N_min=50, œÉ_max=500, D_min=30):
    pixels = image_array.flatten()
    centroids = np.linspace(pixels.min(), pixels.max(), K_init).tolist()

    clusters = {i: [] for i in range(len(centroids))}
    for val in pixels:
        distances = [abs(val - c) for c in centroids]
        closest = np.argmin(distances)
        clusters[closest].append(val)

    clusters = {k: v for k, v in clusters.items() if len(v) >= N_min}

    final_centroids = []
    for values in clusters.values():
        if np.var(values) > œÉ_max:
            sorted_vals = sorted(values)
            mid = len(sorted_vals) // 2
            final_centroids.append(np.mean(sorted_vals[:mid]))
            final_centroids.append(np.mean(sorted_vals[mid:]))
        else:
            final_centroids.append(np.mean(values))

    # Fusion des centres proches
    merged = []
    used = set()
    for i in range(len(final_centroids)):
        if i in used:
            continue
        group = [final_centroids[i]]
        for j in range(i + 1, len(final_centroids)):
            if j not in used and abs(final_centroids[i] - final_centroids[j]) < D_min:
                group.append(final_centroids[j])
                used.add(j)
        merged.append(np.mean(group))
    final_centroids = merged

    # Attribution finale
    segmented = np.zeros_like(pixels)
    labels = np.zeros_like(pixels, dtype=int)
    for i, val in enumerate(pixels):
        distances = [abs(val - c) for c in final_centroids]
        closest = np.argmin(distances)
        segmented[i] = int(final_centroids[closest])
        labels[i] = closest

    return segmented.reshape(image_array.shape), final_centroids, labels

# üî∏ √âvaluation
def intra_cluster_variance(pixels, labels, centroids):
    variance = 0
    for i, c in enumerate(centroids):
        cluster_pixels = pixels[labels == i]
        if len(cluster_pixels) > 0:
            variance += np.var(cluster_pixels)
    return variance / len(centroids)

def mean_centroid_distance(centroids):
    if len(centroids) < 2:
        return 0
    distances = pdist(np.array(centroids).reshape(-1, 1))
    return np.mean(distances)

# üî∏ Traitement de plusieurs images
def process_image(image_path):
    image = Image.open(image_path).convert("L")
    image_array = np.array(image)
    pixels = image_array.flatten().reshape(-1, 1)

    # ISODATA
    segmented_isodata, centroids_iso, labels_iso = isodata_segmentation(image_array)
    intra_iso = intra_cluster_variance(pixels.flatten(), labels_iso, centroids_iso)
    inter_iso = mean_centroid_distance(centroids_iso)

    # K-Means
    kmeans = KMeans(n_clusters=len(centroids_iso), n_init=10, random_state=0)
    labels_km = kmeans.fit_predict(pixels)
    centroids_km = kmeans.cluster_centers_.flatten()
    segmented_kmeans = centroids_km[labels_km].reshape(image_array.shape)
    intra_km = intra_cluster_variance(pixels.flatten(), labels_km, centroids_km)
    inter_km = mean_centroid_distance(centroids_km)

    # Affichage
    plt.figure(figsize=(15, 4))
    plt.suptitle(f"R√©sultats pour : {image_path}", fontsize=14)

    plt.subplot(1, 3, 1)
    plt.title("Originale")
    plt.imshow(image_array, cmap="gray")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("ISODATA")
    plt.imshow(segmented_isodata, cmap="gray")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.title("K-Means")
    plt.imshow(segmented_kmeans, cmap="gray")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    print(f"üìä √âvaluation pour : {image_path}")
    print(f"üîπ ISODATA - Variance intra-cluster : {intra_iso:.3f}, Distance inter-centres : {inter_iso:.2f}")
    print(f"üî∏ K-Means - Variance intra-cluster : {intra_km:.3f}, Distance inter-centres : {inter_km:.2f}")
    print("-" * 60)

# üîª Liste de chemins d‚Äôimages
image_paths = [
    "Untitled.jpeg",  # Remplace avec tes vrais chemins
    "image_euros2.jpg",
    "unnamed.jpg"
]

for path in image_paths:
    process_image(path)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans
from scipy.spatial.distance import pdist

# üîπ Fonction ISODATA simplifi√©e pour images couleur
def isodata_segmentation_color(image_array, K_init=3, N_min=50, œÉ_max=1000, D_min=40):
    pixels = image_array.reshape(-1, 3)
    centroids = np.linspace(pixels.min(), pixels.max(), K_init).reshape(K_init, 1)
    centroids = np.tile(centroids, (1, 3))  # m√™mes valeurs RGB pour d√©marrer

    clusters = {i: [] for i in range(len(centroids))}
    for val in pixels:
        distances = [np.linalg.norm(val - c) for c in centroids]
        closest = np.argmin(distances)
        clusters[closest].append(val)

    clusters = {k: v for k, v in clusters.items() if len(v) >= N_min}

    final_centroids = []
    for values in clusters.values():
        values = np.array(values)
        if np.mean(np.var(values, axis=0)) > œÉ_max:
            sorted_vals = values[values[:, 0].argsort()]  # trier par R
            mid = len(sorted_vals) // 2
            final_centroids.append(np.mean(sorted_vals[:mid], axis=0))
            final_centroids.append(np.mean(sorted_vals[mid:], axis=0))
        else:
            final_centroids.append(np.mean(values, axis=0))

    # Fusion des centres proches
    merged = []
    used = set()
    for i in range(len(final_centroids)):
        if i in used:
            continue
        group = [final_centroids[i]]
        for j in range(i + 1, len(final_centroids)):
            if j not in used and np.linalg.norm(final_centroids[i] - final_centroids[j]) < D_min:
                group.append(final_centroids[j])
                used.add(j)
        merged.append(np.mean(group, axis=0))
    final_centroids = merged

    # Attribution finale
    segmented = np.zeros_like(pixels)
    labels = np.zeros(len(pixels), dtype=int)
    for i, val in enumerate(pixels):
        distances = [np.linalg.norm(val - c) for c in final_centroids]
        closest = np.argmin(distances)
        segmented[i] = final_centroids[closest]
        labels[i] = closest

    return segmented.reshape(image_array.shape).astype(np.uint8), final_centroids, labels

# üî∏ √âvaluation
def intra_cluster_variance(pixels, labels, centroids):
    variance = 0
    for i, c in enumerate(centroids):
        cluster_pixels = pixels[labels == i]
        if len(cluster_pixels) > 0:
            variance += np.mean(np.var(cluster_pixels, axis=0))
    return variance / len(centroids)

def mean_centroid_distance(centroids):
    if len(centroids) < 2:
        return 0
    distances = pdist(np.array(centroids))
    return np.mean(distances)

# üî∏ Traitement d‚Äôimage couleur
def process_image_color(image_path):
    image = Image.open(image_path).convert("RGB")
    image_array = np.array(image)
    pixels = image_array.reshape(-1, 3)

    # ISODATA couleur
    segmented_isodata, centroids_iso, labels_iso = isodata_segmentation_color(image_array)
    intra_iso = intra_cluster_variance(pixels, labels_iso, centroids_iso)
    inter_iso = mean_centroid_distance(centroids_iso)

    # K-Means couleur
    kmeans = KMeans(n_clusters=len(centroids_iso), n_init=10, random_state=0)
    labels_km = kmeans.fit_predict(pixels)
    centroids_km = kmeans.cluster_centers_
    segmented_kmeans = centroids_km[labels_km].reshape(image_array.shape).astype(np.uint8)
    intra_km = intra_cluster_variance(pixels, labels_km, centroids_km)
    inter_km = mean_centroid_distance(centroids_km)

    # Affichage
    plt.figure(figsize=(15, 4))
    plt.suptitle(f"R√©sultats pour : {image_path}", fontsize=14)

    plt.subplot(1, 3, 1)
    plt.title("Originale")
    plt.imshow(image_array)
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("ISODATA Couleur")
    plt.imshow(segmented_isodata)
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.title("K-Means Couleur")
    plt.imshow(segmented_kmeans)
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    print(f"üìä √âvaluation pour : {image_path}")
    print(f"üîπ ISODATA - Variance intra-cluster : {intra_iso:.3f}, Distance inter-centres : {inter_iso:.2f}")
    print(f"üî∏ K-Means - Variance intra-cluster : {intra_km:.3f}, Distance inter-centres : {inter_km:.2f}")
    print("-" * 60)

# üîª Liste d‚Äôimages couleur
image_paths = [
    "Untitled.jpeg",
    "image_euros2.jpg",
    "imagi.jpeg",
    "mad_img.jpeg",
    "regrp.jpeg",
    "reso.jpeg"
]

for path in image_paths:
    process_image_color(path)
```
